```python
from generateIndexFile import get_vol_folder_name_lst
dspath = '../data/PACS/IRB02/'
# dspath = '../data/PACS/Larry-2012-01-17-MRI'
sample_num = 1
if not path.isdir(dspath):
    print("Fail to generateDSIndexFile: dir not exist : "+ dspath)
# return
contents = []
# sort_base = []
vol_ss_keeper = []
vol_cut_group = []

sample_volume = []

for foldername in listdir(dspath):
    folderpath = path.join(dspath, foldername)
    if not path.isdir(folderpath) or len(listdir(folderpath))<10:
        continue
#     print("process: " +folderpath)
    vol_names = get_vol_folder_name_lst(dspath, foldername)
    for vol_name in vol_names:
        vol_path = path.join(dspath, vol_name)
        vd = getVolume(vol_path)
    
        if(vd == None):
            print("==vd invalid==")
            continue
        ss, num_score, tags_score, mask_score = getScore(vd, sample_num)
        vol_ss_keeper.append(ss)
        vol_cut_group.append(vd.meta_data['volume_plane_group'])
        contents.append(get_content_str(vol_name, vd.meta_data) + '\n')
        sample_volume.append(vd.raw_data[:,:,int(vd.raw_data.shape[-1] / 2)])
        

#     sort_base.append(scores[0])
# sorted_list = [x for _, x in sorted(zip(sort_base,contents), key=lambda pair: pair[0], reverse=True)]
# with open(index_file_path, "w") as index_file:
# for line in sorted_list:
#     index_file.write(line)
```

```python
#separate based on cut group
from collections import namedtuple
volumeStruct = namedtuple("volumeStruct", "name sample_img cut_group scores_raw scores_norm")
```

```python
scores_raw = vol_ss_keeper
score_norm = []
scores_np = np.array(scores_raw)
if(scores_np.ndim == 1):
    scores_np = np.expand_dims(scores_np, axis=0)
max_vs = [np.max(scores_np[:,x]) for x in range(scores_np.shape[-1])]
print(max_vs)
for scores in scores_raw:
    score_norm.append([u/v for u,v in zip(scores, max_vs)])
```

```python
vsc = [volumeStruct(name = name_,sample_img = img, cut_group=cg, scores_raw = np.array(ss),scores_norm=None) for name_,img, cg, ss in zip(sample_volume_names, sample_volume, vol_cut_group,vol_ss_keeper)]
```

```python
import matplotlib.pyplot as plt
def imshow_15(sup_title, images, titles):
    fig=plt.figure(figsize=(16, 16))
    columns = 3
    rows = 5
    for i in range(len(images)):
        img = images[i]
        ax=fig.add_subplot(rows, columns, i+1)
        ax.title.set_text(titles[i])#('{:.4f}'.format(titles[i]))
        plt.axis('off')
        plt.imshow(img,cmap='gray')
    fig.suptitle(sup_title, fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
def imshow_25(sup_title, images, titles):
    fig=plt.figure(figsize=(16, 16))
    columns = 5
    rows = 5
    for i in range(len(images)):
        img = images[i]
        ax=fig.add_subplot(rows, columns, i+1)
        ax.title.set_text(titles[i])#('{:.4f}'.format(titles[i]))
        plt.axis('off')
        plt.imshow(img,cmap='gray')
    fig.suptitle(sup_title, fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
#     plt.savefig('save.png')
```
